# Fon Profillerinin Çekimi

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from datetime import date
import warnings


if date.today().day != 15:
    print("Today is not the 15th of the month. Exiting...")
else:
    warnings.filterwarnings("ignore", message="Unverified HTTPS request is being made to host 'www.tefas.gov.tr'.*")
    
    fund_profiles = pd.DataFrame()
    fund_details = pd.DataFrame()
    
    for foncode in islemgoren["Fon Kodu"].unique():
    #for foncode in islemgoren.loc[islemgoren["Fon Kodu"].isin(["LID", "MAC", "AK2"]), "Fon Kodu"]:
    
        
        url = f"https://www.tefas.gov.tr/FonAnaliz.aspx?FonKod={foncode}"
        attempt = 0
        max_attempts = 6 
        max_attempts2 = 10
        
        while True:
            attempt += 1
            
            response = requests.get(url, verify=False)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            try:
                series_element = soup.find('script', string=lambda text: text and 'series' in text).string
                pattern = re.compile(r'"name":"Varlık Dağılımı".*?}', re.DOTALL)
                match = pattern.search(series_element)
                
                if match:
                    varlik_dagilimi_section = match.group()
                    pattern = r'\[([^\[\]]*)\]'
                    matches = re.findall(pattern, varlik_dagilimi_section)
                    result = [eval(match) for match in matches]
                    
                else:
                    print(f"There is negative transaction {foncode}")
                    pattern = re.compile(r'varlık dağılımı.*?series: \[{.*?\}]', re.DOTALL | re.IGNORECASE)
                    match = pattern.search(series_element)
                    varlik_dagilimi_section = match.group()
                    pattern = re.compile(r'\{"name":"(.*?)","id":".*?","data":\[(.*?)\]\}')
                    matches = pattern.findall(varlik_dagilimi_section)
                    result = []
                    for match in matches:
                        category, value = match
                        result.append((category, eval(value)))
    
                
                fund_profiles_temp = pd.DataFrame(result, columns=['Category', 'Value'])
                
                fund_profiles_temp["Fon Kodu"] = foncode
                
                fund_profiles_temp["Tarih"] = date.today()
                
                soup = BeautifulSoup(response.content, 'html.parser')
                table = soup.find('table', {'id': 'MainContent_DetailsViewFund'})
                data = {}
                for row in table.find_all('tr'):
                    # Başlık ve değerleri al
                    header = row.find('td', {'class': 'fund-profile-header'})
                    value = row.find('td', {'class': 'fund-profile-item'})
                    if header and value:
                        data[header.text.strip()] = value.text.strip()
                fund_details_temp = pd.DataFrame([data])
                fund_details_temp["Fon Kodu"] = foncode
                fund_details_temp["Tarih"] = date.today()
                fund_details = pd.concat([fund_details, fund_details_temp])
                fund_profiles = pd.concat([fund_profiles, fund_profiles_temp])
                break  # Exit the retry loop if data fetched successfully
            
            except AttributeError as e:
                if attempt == max_attempts:
                    time.sleep(60)
                elif attempt == max_attempts2:
                    print(f"Max Attempt Reached for {foncode}, BREAKED")
                    break 
                else:
                    print(f"Retrying for {foncode}...")
                    time.sleep(3)  # Wait 3 seconds before retrying
    
    fund_profiles["Value"] = pd.to_numeric(fund_profiles["Value"])
    fund_profiles2 = pd.read_csv("fund_profiles.csv", sep='\t', encoding="utf-8", decimal=',')
    fund_profiles = pd.concat([fund_profiles2,fund_profiles]).reset_index(drop=True)
    fund_profiles.to_csv("fund_profiles.csv", index=False, sep='\t', encoding="utf-8", decimal=',')
    melted_fund_details = pd.melt(fund_details, id_vars=['Fon Kodu',"Tarih"], var_name='Attribute', value_name='Value').sort_values(by="Fon Kodu")
    melted_fund_details2 = pd.read_csv("fund_details.csv", sep='\t', encoding="utf-8", decimal=',')
    melted_fund_details = pd.concat([melted_fund_details2,melted_fund_details]).reset_index(drop=True)
    melted_fund_details.to_csv("fund_details.csv", index=False, sep='\t', encoding="utf-8", decimal=',')
    melted_fund_details
```
