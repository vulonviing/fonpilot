```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Chrome WebDriver'ı başlat
driver = webdriver.Chrome()

# Hedef URL
url = "https://www.tcmb.gov.tr/wps/wcm/connect/TR/TCMB+TR/Main+Menu/Temel+Faaliyetler/Para+Politikasi/Merkez+Bankasi+Faiz+Oranlari/faiz-oranlari"

# Sayfayı aç
driver.get(url)

# Sayfa kaynağını al
page_source = driver.page_source

# BeautifulSoup ile HTML analizi
soup = BeautifulSoup(page_source, 'html.parser')

# Tabloyu bul
table = soup.find('table', {'class': 'table-condensed table-striped borderless small-text table'})

# Tabloyu çek
data = []
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all(['td', 'th'])
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# DataFrame'e dönüştür
df = pd.DataFrame(data[1:], columns=data[0])

# Chrome'u kapat
driver.quit()

# DataFrame'i yazdır
print(df)
```

```python
df["Tarih"] = pd.to_datetime(df["Tarih"])
# Eksik tarihleri oluştur
start_date = df['Tarih'].min()
end_date = df['Tarih'].max()
date_range = pd.date_range(start_date, end_date)

# Eksik tarihleri içeren yeni DataFrame oluştur
df_all_dates = pd.DataFrame({'Tarih': date_range})

# Orijinal DataFrame ile eksik tarihleri içeren DataFrame'i birleştir
df_merged = pd.merge(df_all_dates, df, on='Tarih', how='left')

# 'Borç Alma' ve 'Borç Verme' sütunlarındaki eksik değerleri doldur
df_merged['Borç Alma'].fillna(method='ffill', inplace=True)
df_merged['Borç Verme'].fillna(method='ffill', inplace=True)
df_merged["Borç Alma"] = pd.to_numeric(df_merged["Borç Alma"])
df_merged["Borç Verme"] = pd.to_numeric(df_merged["Borç Verme"])

df_merged["faiz_gunluk"] = (df_merged["Borç Verme"]/365) * 0.01
df_merged["Tarih"] = pd.to_datetime(df_merged["Tarih"])

df_merged["Year"] = df_merged["Tarih"].dt.year
df_merged['count'] = df_merged.groupby(['Year', 'faiz_gunluk']).cumcount() + 1
df_merged["faiz"] = df_merged["faiz_gunluk"] * df_merged["count"]

df_merged.to_excel("mevduat.xlsx")
df_merged
```
