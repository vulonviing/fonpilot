# Teknoloji Teması

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the website
website_url = 'https://fintables.com/fonlar/IRT/rakip-analizi'
driver.get(website_url)

# Wait for the page to load (you may need to adjust the sleep time accordingly)
import time
time.sleep(5)

# Get the page source
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Extract fund codes from the HTML
fund_codes = []
fund_elements = soup.select('tr.cursor-pointer a[href^="/fonlar/"]')
for fund_element in fund_elements:
    fund_code = fund_element['href'].split('/')[-1]
    fund_codes.append(fund_code)

# Create a DataFrame
df = pd.DataFrame({'Fund Codes': fund_codes})

# Save the DataFrame to a CSV file


df["Tema"] = "Teknoloji"
sa = pd.concat([sa, df])
sa
```

# Sürdürülebilirlik

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the website
website_url = 'https://fintables.com/fonlar/MTS/rakip-analizi/surdurulebilirlik'
driver.get(website_url)

# Wait for the page to load (you may need to adjust the sleep time accordingly)
import time
time.sleep(5)

# Get the page source
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Extract fund codes from the HTML
fund_codes = []
fund_elements = soup.select('tr.cursor-pointer a[href^="/fonlar/"]')
for fund_element in fund_elements:
    fund_code = fund_element['href'].split('/')[-1]
    fund_codes.append(fund_code)

# Create a DataFrame
df = pd.DataFrame({'Fund Codes': fund_codes})

# Save the DataFrame to a CSV file


df["Tema"] = "Sürdürülebilirlik"
sa = pd.concat([sa, df])
sa
```

```python
sa.reset_index(inplace=True, drop=True)

sa
```

```python
def join_themes(themes):
    return ', '.join(set(themes))

# Group by 'Codes' and aggregate 'Tema' using the custom function
sa = sa.groupby('Fund Codes')['Tema'].agg(join_themes).reset_index()

sa
```

```python
sa.loc[sa["Tema"].str.contains("Teknoloji")]
sa.to_csv("fon_temalari.csv", index=False)
```
